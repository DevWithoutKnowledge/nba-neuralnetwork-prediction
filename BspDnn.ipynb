```python
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import classification_report
from tensorflow.keras import backend as K
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras.regularizers import l2
from tensorflow.keras.metrics import AUC
from tensorflow.keras import activations

#datasets
trainX = train_df
trainY = y_train
testX = test_df
testY = y_test

          
# define the architecture using Keras
batch = 32
model = tf.keras.Sequential([
    tf.keras.layers.Dense(200, input_shape=(14,12),kernel_regularizer=l2(0.01)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Activation(activations.relu),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(100,kernel_regularizer=l2(0.01)),
    # tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Activation(activations.relu),
    # tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(40,kernel_regularizer=l2(0.005)),
    # tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Activation(activations.relu),
    # tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

class_weight = {0: 1.1,
                1: 1.0}

# train the model using Adam
print("[INFO] training network...")
opt = tf.keras.optimizers.Adam(learning_rate = 0.001,clipvalue=1)
lossFunction = tf.keras.losses.BinaryCrossentropy(from_logits=True) #from_logits=True
model.compile(loss=lossFunction,optimizer=opt,metrics=[tf.keras.metrics.BinaryAccuracy(threshold = 0.50),tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), AUC(name='auc')])
history = model.fit(trainX, trainY, validation_data = (testX, testY), shuffle=True, epochs = 60, batch_size = batch, class_weight=class_weight)

# plot
plt.plot(history.history['loss'], label = 'loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.ylim([0.5, 0.7])
plt.title('training and validation loss')
plt.xlabel('loss')
plt.ylabel('Error')
plt.legend()
plt.show()

plt.plot(history.history['auc'], label = 'auc')
plt.plot(history.history['val_auc'], label = 'val_auc')
plt.ylim([0.5, 0.7])
plt.title('training and validation AUC')
plt.xlabel('auc')
plt.ylabel('value')
plt.legend()
plt.show()

plt.plot(history.history['binary_accuracy'], label = 'accuracy')
plt.plot(history.history['val_binary_accuracy'], label = 'val_accuracy')
plt.ylim([0.5, 0.7])
plt.title('training and validation Acc')
plt.xlabel('acc')
plt.ylabel('value')
plt.legend()
plt.show()
```
